```python
import pandas as pd
import numpy as np


def auction_bids(game_info, current_info, last_auction_result, player_information):
    round_number = game_info['Round_Number']
    total_items = game_info['Total_Number_Items']
    robots = game_info['Number_Robots']
    remaining_items = current_info['Remaining_Number_Items']
    current_vp = current_info['vp']
    current_budget = current_info['remaining_budget']
    last_winner = last_auction_result['winner']
    last_winning_bid = last_auction_result['winning_bid']
    last_vp = last_auction_result['vp']
    last_total_bid = last_auction_result['total']
    last_min_bid = last_auction_result['min']
    robots_name = ['Robot_' + str(i+1) for i in range(robots)]

    current_auction = total_items - remaining_items + 1

    # Record the player_info
    if current_auction == 1:
        for i in robots_name:
            player_information[i] = {'total_vp': 0, 'max_budget': 1}
    else:
        if last_winner != 'player':
            for i in robots_name:
                if last_winner == i:
                    player_information[i]['total_vp'] += last_vp
                    player_information[i]['max_budget'] -= last_winning_bid
                else:
                    player_information[i]['max_budget'] -= last_min_bid

    # Use the bid/VP ratio to determine the bid value
    if current_auction == 1:
        bid_value = 0
        return bid_value

    ratio = last_winning_bid / last_vp

    if current_budget > 0.3:
        bid_value = min(ratio * current_vp, current_budget)
    else:
        bid_value = 0

    return bid_value
```
